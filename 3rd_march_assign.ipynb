```python
import seaborn as sns
```

```python
"""Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

    Ans: 1. Scatter plot: Scatter plot is used to visualize the relationship between two variables in a dataset. 
            It is useful to identify patterns, trends, and outliers in the data. For example, it can be used to plot the relationship between the height and weight of individuals.
         2. Line plot: Line plot is used to visualize trends over time or any continuous variable. It can be used to plot time 
            series data or data with a continuous X-axis. For example, it can be used to plot the temperature changes over a year.
         3. Bar plot: Bar plot is used to visualize categorical data with numerical values. It can be used to compare the values 
            of different categories. For example, it can be used to compare the sales of different products.
         4. Histogram: Histogram is used to visualize the distribution of a numerical variable. It can be used to identify 
            the skewness, central tendency, and spread of the data. For example, it can be used to plot the distribution of 
            the age of a population.
         5. Heatmap: Heatmap is used to visualize the correlation between variables in a dataset. It can be used to identify 
            patterns and relationships between variables. For example, it can be used to plot the correlation between the different features of a dataset.
"""
```

```python
"""Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset."""

fmri = sns.load_dataset('fmri')

x = fmri['timepoint']
y = fmri['signal']

sns.lineplot(x = x, y = y, hue = 'event', style = 'region', data = fmri)
```

```python
"""Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset."""

titanic = sns.load_dataset('titanic')
x1 = titanic['pclass']
y1 = titanic['age']
y2 = titanic['fare']
sns.boxplot(x = x1, y = y1)
sns.boxplot(x = x1, y = y2)
```

```python
"""Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset."""

diamonds = sns.load_dataset('diamonds')
sns.histplot(data = diamonds, x = "price", hue = "cut", kde = True)
```

```python
"""Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset."""

iris = sns.load_dataset('iris')
sns.pairplot(data = iris, hue = 'species')
```

```python
"""Que 6: Use the "flights" dataset from seaborn to plot a heatmap."""

flights = sns.load_dataset('flights')
data = flights.pivot('month', 'year', 'passengers')
sns.heatmap(data)
```
